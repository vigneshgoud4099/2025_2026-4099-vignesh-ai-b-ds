```
import pandas as pd

# Iris Dataset URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

df = pd.read_csv(url)

print("Original dataset:")
print(df.head())

# AND condition
print("\nRecords where sepal_length > 5 and petal_length > 4:")
filtered_and = df[(df['sepal_length'] > 5) & (df['petal_length'] > 4)]
print(filtered_and)

# OR condition
print("\nRecords where species is setosa or virginica:")
filtered_or = df[(df['species'] == 'setosa') | (df['species'] == 'virginica')]
print(filtered_or)

# NOT condition
print("\nRecords where species is NOT setosa:")
filtered_not = df[~(df['species'] == 'setosa')]
print(filtered_not)

# Combined condition
print("\nSpecies is versicolor and sepal_width > 3 and petal_length > 4:")
combined = df[(df['species'] == 'versicolor') &
              (df['sepal_width'] > 3) &
              (df['petal_length'] > 4)]
print(combined)

print("\nNumber of records satisfying combined condition:", len(combined))

# Sorting
print("\nSorted by sepal_length (Ascending):")
sorted_df = df.sort_values(by='sepal_length')
print(sorted_df.head())

print("\nSorted by petal_length (Descending):")
sorted_desc = df.sort_values(by='petal_length', ascending=False)
print(sorted_desc.head())

print("\nSorted by species and sepal_length:")
sorted_multi = df.sort_values(by=['species', 'sepal_length'])
print(sorted_multi.head())

# Groupby operations
print("\nAverage sepal_length by species:")
avg_sepal = df.groupby('species')['sepal_length'].mean()
print(avg_sepal)

print("\nAverage petal_length by species:")
avg_petal = df.groupby('species')['petal_length'].mean()
print(avg_petal)

print("\nCount of records by species:")
count_species = df['species'].value_counts()
print(count_species)
```
