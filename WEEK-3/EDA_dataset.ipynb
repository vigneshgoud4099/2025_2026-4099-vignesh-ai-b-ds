```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# New Dataset URL
url = "https://raw.githubusercontent.com/plotly/datasets/master/winequality-red.csv"

print("Dataset:")
df = pd.read_csv(url)
print(df.head())

# Select a numeric column
data = df['alcohol']

print("\nSummary statistics:")
print(data.describe())

print("\nVariance:", np.var(data))
print("Standard deviation:", np.std(data))

# Mean, Median, Mode
mean_value = np.mean(data)
median_value = np.median(data)
mode_value = data.mode()

# Histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')

plt.axvline(mean_value, color='red', linestyle='dashed',
            linewidth=2, label=f"Mean: {mean_value:.2f}")

plt.axvline(median_value, color='green', linestyle='dashed',
            linewidth=2, label=f"Median: {median_value:.2f}")

plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.title('Distribution of Alcohol Content')
plt.legend()
plt.show()

print("\nInference:")
print("Mean:", mean_value)
print("Median:", median_value)

if len(mode_value) == 1:
    print("Mode:", mode_value[0])
else:
    print("Mode:", list(mode_value))

if mean_value > median_value:
    print("The distribution is positively skewed.")
elif mean_value < median_value:
    print("The distribution is negatively skewed.")
else:
    print("The distribution is symmetric.")
```
