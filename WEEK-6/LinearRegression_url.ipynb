```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
url="https://raw.githubusercontent.com/ybiFoundation/Dataset/main/Salary%20Data.csv"
data = pd.read_csv(url)
print("Dataset preview")
print(data.head())
x = data[['Experience Years']]
y=data['Salary']
model=LinearRegression()
model.fit(x,y)
print("\nModel parameter")
print("Stop",model.coef_[0])
print("Intercept:",model.intercept_)
Years=[[5]]
predicted_salary=model.predict(Years)
print("\n predicted salary for 5 Years experience:")
y_pred=model.predict(x)
mse=mean_squared_error(y,y_pred)
print("\n Mean Squared Error",mse)
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.title("LinearRegression Salary prediction")
plt.show()
```
