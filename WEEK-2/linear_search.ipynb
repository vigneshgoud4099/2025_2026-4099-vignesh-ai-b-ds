```
#linear search
n=int(input("enter the number of elements:"))
arr=[]
for i in range(n):
  arr.append(int(input(f"enter the element {i+1}:")))
key=int(input("enter the elemnet to search:"))
found=False
for i in range(n):
  if arr[i]==key:
    print(f"element {key} found at position {i+1}")
    found=True
    break
if not found:
  print("elemnet {key} not found")
```
