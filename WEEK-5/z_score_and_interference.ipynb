```

#calculating z-score
import numpy as np
data=[10,20,30,40,50]
mean=np.mean(data)
print("mean:",mean)
std_dev=np.std(data,ddof=1)
print("standard deviation:",std_dev)
x=40
z_score=(x-mean)/std_dev
print("z score of 40:",z_score)
z_scores=[(x-mean)/std_dev for x in data]
print("z_scores:",z_scores)
```
